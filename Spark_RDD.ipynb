```
from pyspark.sql import SparkSession
from functools import reduce
from pyspark.sql.functions import lower,col

# Create SparkSession object
spark = SparkSession.builder \
                    .master("local[*]") \
                    .appName("Spark Assignment") \
                    .getOrCreate()


#spark version
print("Spark version is "+ str(spark.version))

```

```
customer = spark.read.csv("/FileStore/tables/cust_dimen.csv",
                         sep=',',
                         header=True,
                         inferSchema=True,
                         nullValue='NA')
# View the first five records
customer.show(5)
print("After Applying Count on RDD:-")
#Get the number of records
print("The CUSTOMER RDD contains %d records." % customer.count())
reduce_rdd = (reduce(
    lambda memo_df, col_name: memo_df.withColumn(col_name, lower(col(col_name))),
    customer.columns,
    customer
))
print("After Applying Reduce on RDD:-")
reduce_rdd.show(5)
```
